```python
%load_ext autoreload
%autoreload 2
%matplotlib inline

from scipy.io import arff
import numpy as np
from ant_cluster import AntCluster
import pylab
```

```python
# data = np.load("data/just_locations.npy")
data = np.load("data/locations_and_word_indexes.npy")  # location, idx
# np.set_printoptions(threshold=np.nan)
# np.set_printoptions(formatter={'all':lambda x: str(x)})
print(data.dtype)
# np.set_printoptions(threshold=np.inf)
print(data)
```

# Experiment with mixed matrices

```python
ac = AntCluster(data, mask=["cosine", "discrete"], option="mixed")
# runs = len(labels) * 100
ac.similate(data.shape[0] * 100)
```

```python
ant_groups = ac.group()
# print(ant_groups)
# ant_groups.sort(key=lambda ag: ag.gene[0], reverse=True)
for i, ant_group in enumerate(ac.group()):
#     sorted_ant_group = ant_group
    ant_group.sort(key=lambda ant: ant.gene[0], reverse=True)
    for ant in ant_group:
        print(ant.gene, f"cluster {i}")
```

# Experiment with truncation

```python
sort_data = data[data[:,0].argsort()]
buck_data = np.array_split(sort_data, 150)
trunc_buck_data = [np.squeeze(buck[:, 1]).astype(int) for buck in buck_data]
ants = np.array([np.sum(np.eye(158)[idx_lst], 0) for idx_lst in trunc_buck_data])
ants.shape, ants
```

```python
ac = AntCluster(ants, label=np.arange(0, ants.shape[0]), option="cosine")
ac.similate(ants.shape[0] * 1000)
```

```python
group = ac.group()

for i, ant_group in enumerate(ac.group()):
    print(f"cluster {i}")
    for ant in ant_group:
        print(ant.true_label)
```

```python
labels = ["create nest", "join nest", "positive meeting", "negative meeting", "different nest meeting"]
trace = np.array(ac.statistic).T
for i in range(trace.shape[0]):
    pylab.plot(trace[i,:].cumsum(), label=labels[i])
    
# pylab.yscale('symlog', linthreshy=0.015)
pylab.legend()
pylab.grid()
```

```python
X = lda.datasets.load_reuters()
vocab = lda.datasets.load_reuters_vocab()
titles = lda.datasets.load_reuters_titles()

model = lda.LDA(n_topics=20, n_iter=1500, random_state=1)
model.fit(X)  # model.fit_transform(X) is also available
topic_word = model.topic_word_  # model.components_ also works
n_top_words = 8
```

```python
for i, topic_dist in enumerate(topic_word):
    topic_words = np.array(vocab)[np.argsort(topic_dist)][:-(n_top_words+1):-1]
    print('Topic {}: {}'.format(i, ' '.join(topic_words)))
```
